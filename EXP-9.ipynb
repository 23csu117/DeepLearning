# Experiment 9: CNN Model for CIFAR-10 Image Classification

**Objective:** Train a Convolutional Neural Network (CNN) model to classify images from the CIFAR-10 dataset.

**Libraries Used:** NumPy, Matplotlib, TensorFlow / Keras, scikit-learn


## 1. Introduction

CIFAR-10 is a popular image classification dataset containing 60,000 color images of size 32×32 across 10 classes: airplane, automobile, bird, cat, deer, dog, frog, horse, ship, and truck.

In this experiment, we build and train a CNN model using Keras to classify images from CIFAR-10.


## 2. Aim of the Experiment

In this notebook, we will:

1. Load the CIFAR-10 dataset
2. Preprocess the images and labels
3. Build a CNN model
4. Train the model
5. Evaluate test accuracy
6. Visualize predictions and performance


```python
# Install if needed
# !pip install tensorflow matplotlib numpy scikit-learn
```

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf

from tensorflow.keras.datasets import cifar10
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization
from tensorflow.keras.utils import to_categorical
from sklearn.metrics import confusion_matrix, classification_report

np.random.seed(42)
tf.random.set_seed(42)
```

## 3. Load the CIFAR-10 Dataset


```python
(x_train, y_train), (x_test, y_test) = cifar10.load_data()

print('Training images shape:', x_train.shape)
print('Training labels shape:', y_train.shape)
print('Testing images shape:', x_test.shape)
print('Testing labels shape:', y_test.shape)
```

```python
class_names = ['Airplane', 'Automobile', 'Bird', 'Cat', 'Deer', 'Dog', 'Frog', 'Horse', 'Ship', 'Truck']

plt.figure(figsize=(12, 5))
for i in range(10):
    plt.subplot(2, 5, i + 1)
    plt.imshow(x_train[i])
    plt.title(class_names[int(y_train[i])])
    plt.axis('off')
plt.tight_layout()
plt.show()
```

## 4. Data Preprocessing

We normalize pixel values to the range [0, 1] and convert class labels into one-hot encoded vectors.


```python
x_train = x_train.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0

y_train_cat = to_categorical(y_train, 10)
y_test_cat = to_categorical(y_test, 10)

print('Normalized training image shape:', x_train.shape)
print('One-hot encoded labels shape:', y_train_cat.shape)
```

## 5. Build the CNN Model


```python
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', padding='same', input_shape=(32, 32, 3)),
    BatchNormalization(),
    Conv2D(32, (3, 3), activation='relu', padding='same'),
    MaxPooling2D((2, 2)),
    Dropout(0.25),

    Conv2D(64, (3, 3), activation='relu', padding='same'),
    BatchNormalization(),
    Conv2D(64, (3, 3), activation='relu', padding='same'),
    MaxPooling2D((2, 2)),
    Dropout(0.25),

    Conv2D(128, (3, 3), activation='relu', padding='same'),
    BatchNormalization(),
    Conv2D(128, (3, 3), activation='relu', padding='same'),
    MaxPooling2D((2, 2)),
    Dropout(0.25),

    Flatten(),
    Dense(256, activation='relu'),
    Dropout(0.5),
    Dense(10, activation='softmax')
])

model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

model.summary()
```

## 6. Train the Model


```python
history = model.fit(
    x_train, y_train_cat,
    epochs=15,
    batch_size=128,
    validation_split=0.1,
    verbose=1
)
```

## 7. Plot Accuracy and Loss Curves


```python
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Training and Validation Accuracy')
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training and Validation Loss')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

## 8. Evaluate the Model on Test Data


```python
test_loss, test_accuracy = model.evaluate(x_test, y_test_cat, verbose=0)
print('Test Loss:', round(test_loss, 4))
print('Test Accuracy:', round(test_accuracy, 4))
```

## 9. Make Predictions


```python
y_pred_probs = model.predict(x_test, verbose=0)
y_pred = np.argmax(y_pred_probs, axis=1)
y_true = y_test.flatten()

print('Sample predicted classes:', y_pred[:10])
```

```python
plt.figure(figsize=(12, 6))
for i in range(10):
    plt.subplot(2, 5, i + 1)
    plt.imshow(x_test[i])
    plt.title(f'True: {class_names[y_true[i]]}\nPred: {class_names[y_pred[i]]}')
    plt.axis('off')
plt.tight_layout()
plt.show()
```

## 10. Confusion Matrix and Classification Report


```python
cm = confusion_matrix(y_true, y_pred)
print('Confusion Matrix:\n')
print(cm)
```

```python
print('Classification Report:\n')
print(classification_report(y_true, y_pred, target_names=class_names))
```

## 11. Conclusion

In this experiment, we built and trained a CNN model for CIFAR-10 image classification. The model learned hierarchical visual features from color images and classified them into 10 object categories. This experiment demonstrates the usefulness of CNNs for real-world image recognition tasks.
